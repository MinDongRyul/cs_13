### 2번 문제

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/healthexp.csv")
df
```

```python
df[df['Year'] >= 2011].pivot_table(index='Year', 
                                   values='Life_Expectancy',
                                   columns='Country',
                                   aggfunc='mean')
```
